```python
import pandas as pd
from sqlalchemy import create_engine

# -----------------------------
# MySQL Database Connection
# -----------------------------
username = "username (usally root)"
password = "yourpassword"
host = "localhost:portnumber"
database = "Olist" 

engine = create_engine(f"mysql+pymysql://{username}:{password}@{host}/{database}")

# -----------------------------
# CSV paths
# -----------------------------
tables_to_load = {
    "customers": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_customers_dataset.csv",
    "orders": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_orders_dataset.csv",
    "order_items": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_order_items_dataset.csv",
    "products": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_products_dataset.csv",
    "payments": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_order_payments_dataset.csv",
    "sellers": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_sellers_dataset.csv",
    "reviews": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\olist_order_reviews_dataset.csv",
    "category_translation": r"C:\Users\User\OneDrive\Documents\Ecomm_project_DA\Project Files\product_category_name_translation.csv"
}

# -----------------------------
# REQUIRED COLUMNS per table
# -----------------------------
required_columns = {
    "customers": ["customer_id","customer_unique_id" ,"customer_city"],
    "orders": ["order_id", "customer_id", "order_purchase_timestamp","order_delivered_customer_date"],
    "order_items": ["order_id", "product_id", "price"],
    "products": ["product_id", "product_category_name"],
    "payments": ["order_id", "payment_type", "payment_value"],
    "sellers": ["seller_id", "seller_city"],
    "reviews": ["order_id", "review_score"],
    "category_translation": ["product_category_name", "product_category_name_english"]
}

# -----------------------------
# Load only required columns
# -----------------------------
for table_name, csv_file in tables_to_load.items():
    print(f"\nLoading -> {table_name}")

    df = pd.read_csv(csv_file)

    # Filter required columns
    df = df[required_columns[table_name]]

    df.to_sql(table_name, con=engine, if_exists="replace", index=False)

    print(f"{table_name} loaded successfully with selected columns.")

print("\nAll tables loaded with required columns only!")
```
